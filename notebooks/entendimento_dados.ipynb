# Entendimento dos dados

```python
import pandas as pd
```

## Atividades

```python
df_atividades = pd.read_excel('data/atividades.xlsx')
df_atividades.head()
```

```python
pd.DataFrame({
    'Tipo de dado': df_atividades.dtypes,
    'Quantidade de nulos': df_atividades.isnull().sum(),
    'Percentual de nulos': df_atividades.isnull().mean()*100,
    'Qtde valores únicos': df_atividades.nunique(),
    'Exemplos valores únicos': df_atividades.apply(lambda x: x.unique()[:5])
})
```

```python
df_atividades.shape
```

```python
df_atividades.describe()
```

Nenhum nulo ou valor discrepante.
Coluna PK só contém valores únicos.

```python
df_atividades.to_csv('data/atividades.csv', index=False)
```
